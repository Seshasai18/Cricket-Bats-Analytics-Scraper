```python
import pandas as pd
import requests
from bs4 import BeautifulSoup as bs

data = []
search_ele= 'condor'
url = f"https://www.dsc-cricket.com/catalogsearch/result/?q={search_ele}"
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
                  "AppleWebKit/537.36 (KHTML, like Gecko) "
                  "Chrome/120.0.0.0 Safari/537.36"
}

web_page = requests.get(url, headers=headers)
web_content = bs(web_page.text, "html.parser")
products = web_content.find_all('li',{'class':"item product product-item testone"})
product_name = 'product'
for product in products:

    try:
        product_name = product.find(
            'div', class_="product name product-item-name"
        ).a.text.strip()
    except:
        product_name = "No product name"

    try:
        price = product.find('span', class_='price').text.strip()
    except:
        price = "Not available"

    try:
        original_price = product.find(
            'span', class_="price-container price-final_price tax weee price-default"
        ).span.text.strip()
    except:
        original_price = "Not available"

    try:
        discount = product.find(
            'span', class_="percent-off-main 222"
        ).get_text(strip=True)
    except:
        discount = "Not available"

    try:
        img_link = product.find('img')['src']
    except:
        img_link = "Not available"

    product_link = "Not available"
    product_short_description = "Not available"
    product_long_description = "Not available"
    product_weight = "Not available"

    try:
        product_link = product.find(
            'div', class_="product name product-item-name"
        ).a['href']

        headers = {
            "User-Agent": "Mozilla/5.0"
        }

        product_page = requests.get(product_link, headers=headers)
        product_content = bs(product_page.text, "html.parser")

        try:
            product_short_description = product_content.find(
                'div', class_="short_description_pr"
            ).h3.text.strip()
        except:
            pass

        try:
            product_long_description = product_content.find(
                'div', class_="short_description_pr"
            ).p.text.strip()
        except:
            pass

        try:
            product_weight = product_content.find(
                'div', class_="product attribute description"
            ).find_all('li')[0].text.strip()
        except:
            pass

    except:
        pass
    data.append({
        "Product Name": product_name,
        "Price": price,
        "Original Price": original_price,
        "Discount": discount,
        "Image Link": img_link,
        "Product Link": product_link,
        "Short Description": product_short_description,
        "Long Description": product_long_description,
        "Weight": product_weight
    })
df = pd.DataFrame(data)
print(df.head())
df.to_csv(f"{search_ele}.csv", index=False)
print("Data saved successfully")
```
